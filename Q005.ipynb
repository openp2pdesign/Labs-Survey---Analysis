# Q005 - Quanto è grande la sede del laboratorio?

```python
# -*- coding: UTF-8 -*-

# Render our plots inline
%matplotlib inline 

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn

pd.set_option('display.mpl_style', 'default') # Make the graphs a bit prettier, overridden by seaborn
pd.set_option('display.max_columns', None) # Display all the columns
plt.rcParams['font.family'] = 'sans-serif' # Sans Serif fonts for all the graphs

# Reference for color palettes: http://web.stanford.edu/~mwaskom/software/seaborn/tutorial/color_palettes.html
```

```python
# Load csv file first
data = pd.read_csv("data/lab-survey.csv", encoding="utf-8")
```

```python
# Check data
#data[0:4] # Equals to data.head()
```

```python
# Get the distribution of ages
space = data["D5"].value_counts()
print "Data:"
print space
print ""
print "Data %:"
print data["D5"].value_counts(normalize=True) * 100
print ""
print "Data: statistics:"
print data["D5"].describe()

# Plot the data 01
plt.figure(figsize=(8,6))
plt.xlabel(u'Dimensione spazio (mq)')
plt.ylabel('Lab')
my_colors = seaborn.color_palette("husl", len(space)) # Set color palette
space.plot(kind="bar",title=u"Quanto è grande la sede del laboratorio?",color=my_colors)
plt.savefig(u"pdf/Q005-DimensioneLab01.pdf") # Save pdf
```

```python
# Plot the data 02

# Reorder value_counts by index (age) natural order
space1 = space.sort_index()

plt.figure(figsize=(8,6))
plt.title(u'Quanto è grande la sede del laboratorio?')
plt.xlabel(u'Dimensione spazio (mq)')
plt.ylabel('Lab')

# Plot the data
my_colors = seaborn.color_palette("husl", len(space1)) # Set color palette
space1.plot(kind='bar',title=u"Quanto è grande la sede del laboratorio?",color=my_colors)
plt.savefig(u"pdf/Q005-DimensioniLab02.pdf") # Save pdf
```

```python
# Check histogram
plt.figure(figsize=(8,6))
plt.title(u'Quanto è grande la sede del laboratorio?')
plt.xlabel(u'Dimensioni (mq)')
plt.ylabel('Lab')
data["D5"].hist(bins=60)
plt.savefig(u"pdf/Q005-DimensioniLab03.pdf") # Save pdf
```
