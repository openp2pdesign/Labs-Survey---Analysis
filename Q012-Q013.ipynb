```python
# From http://stackoverflow.com/questions/7684475/plotting-labeled-intervals-in-matplotlib-gnuplot

from matplotlib.dates import DateFormatter, MinuteLocator, SecondLocator
import datetime as dt
from pylab import rcParams


#Plot function
def timelines(y, xstart, xstop, color='b'):  
    plt.hlines(y, xstart, xstop, color, lw=4)
    plt.vlines(xstart, y+0.03, y-0.03, color, lw=2)
    plt.vlines(xstop, y+0.03, y-0.03, color, lw=2)

for i in history.index:
    timelines(history["Row"][i], pd.to_datetime(history["D12"][i]), pd.to_datetime(history["D13"][i]), 'r')

#Setup the plot
rcParams['figure.figsize'] = 10, 20
ax = plt.gca()
ax.xaxis_date()
myFmt = DateFormatter('%Y-%m-%d')
ax.xaxis.set_major_formatter(myFmt)
#ax.xaxis.set_major_locator(SecondLocator(interval=20)) # used to be SecondLocator(0, interval=20)

ax.yaxis.set_ticks(range(len(categories)))
ax.yaxis.set_ticklabels(categories)
#plt.xticks(rotation=70)
plt.xlim(dStart,dEnd)
plt.xlabel('Time')
plt.savefig("svg/Q012-13-PeriodoDiSviluppo.svg")
plt.savefig("png/Q012-13-PeriodoDiSviluppo.png")
plt.show()
```
