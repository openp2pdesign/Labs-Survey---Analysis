```python
# Set the main figure
fig = plt.figure(figsize=(28,6))
fig.suptitle(u"Quando è aperto al pubblico il laboratorio?", fontsize=18, y=1.08)

# Set the subplots
l = 0
ax= []
for i in (range(7)):
    for k in (range(2)):
        l+=1
        ax.append(fig.add_subplot(2,7,l))
        ax[l-1].set_xlabel('Apertura')
        ax[l-1].set_ylabel('Lab')
        ax[l-1].set_title(subquestions_value[l-1])

# Set the right margins
fig.tight_layout()



#my_colors = seaborn.color_palette("husl", len(space)) # Set color palette
#space.plot(kind="bar",color=my_colors)

plt.savefig(u"svg/Q027-AperturaLaboratorio.svg")
plt.savefig(u"png/Q027-AperturaLaboratorio.png")
```
